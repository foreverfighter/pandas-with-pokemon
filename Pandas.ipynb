### Reorder columns

```python
cols = df.columns.tolist()
cols.remove('name')
cols.insert(0, 'name')
df = df[cols]
```

```python
df.head(1)
```

### Filter bug type pokemon

```python
df[df.type1 == 'bug']
```

```python
df[(df.type1 == 'bug') & (df.sp_attack > 40)]
```

### Create new col with values based on other columns

This creates a new column `newcol` and sets the value to 'super' for rows where the `against_bug` value is 2.0. 

```python
df.loc[df.against_bug == 2.0, 'newcol'] = 'super'
```

```python
df[df.newcol=='super']
```

```python
df.newcol
```

### Fill NA values

```python
df.newcol = df.newcol.fillna('not super')
```

### Save as csv
